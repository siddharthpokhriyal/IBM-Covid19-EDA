```python
df3=df.drop(['Time','ConfirmedForeignNational','ConfirmedIndianNational'],axis=1)
```

```python
df3.head()
```

```python
# Finding active cases
df3['Active']=df3['Confirmed']-(df3['Cured']+df3['Deaths'])
```

```python
df3.head()
```

```python
df3.tail(20)
```

```python
# covid data till 7 july 2021
df4=df3.loc[df['Date']=='2021-07-07']
```

```python
df4=df4.sort_values('Confirmed')
```

```python
df4
```

```python
ax = df4.plot(x="State/UnionTerritory", y=['Confirmed','Cured'],kind='bar')
ax.get_yaxis().get_major_formatter().set_scientific(False)
plt.ylabel('CASES')
plt.rcParams['figure.figsize'] = [20, 15]
plt.show()
```

#### The above plot shows Andhra Pradesh, Tamil Nadu, Karnataka, Kerala, Maharashtra are top 5 states with higest confirmed cases and heighest cured cases till 7 july 2021

```python
# Total deaths till 7 july 2021 state wise
sns.set_theme(style="whitegrid")
ax = sns.barplot(x="State/UnionTerritory", y="Deaths", data=df4)
plt.xticks(rotation=90)
plt.show()
```

### Top 10 states with highest confirmed cases for active cases

```python
# Date wise analysis of top 10 states with highest confirmed cases for active cases
Top_10_states=df4.iloc[-10:,1]
fig=plt.figure(figsize=(15,8))
ax=sns.lineplot(data=df3[df3['State/UnionTerritory'].isin
                                (Top_10_states)],x='Date',y='Active',hue='State/UnionTerritory')
ax.set_title("Active cases by States", size=20)

plt.legend(bbox_to_anchor=(1.02, 1), loc=2)
plt.show()
```

## Vaccine Analysis

```python
df6=pd.read_csv('covid_vaccine_statewise.csv')
df6
```

```python
df6.info()
```

```python
df7=df6[df6['Total Sessions Conducted'].notna()]
```

```python
df7.info()
```

```python
df7.isnull().sum()
```

```python
# droping column with most null values
df8 = df7.drop(['Total Sputnik V Administered','AEFI', '18-45 years (Age)','45-60 years (Age)','60+ years (Age)'], axis=1)
```

```python
df8.info()
```

## Vaccination Analysis inIndia

```python
df8[df8['State']=='India'].tail()
```

```python
# gender ratio for vaccination
male=df8[df8['State']=='India']['Male(Individuals Vaccinated)'].max()
female=df8[df8['State']=='India']['Female(Individuals Vaccinated)'].max()
transgender=df8[df8['State']=='India']['Transgender(Individuals Vaccinated)'].max()
labels = ['Male','Female','Transgender']
sizes = [male,female,transgender]
#add colors
colors = ['#ff9999','#66b3ff','#99ff99']
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, colors=colors, autopct='%1.3f%%',
         startangle=90)
plt.legend(title="Vaccinated ratio for Covid19",bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#doses administrated vs Indivisuals valccinated
Doses = df8[df8['State']=='India']["Total Doses Administered"].max() 
Vaccinated = df8[df8['State']=='India']["Total Individuals Vaccinated"].max() 
labels = ['Doses Administered','Individuals Vaccinated']
sizes = [Doses,Vaccinated]
colors = ['#66b3ff','#99ff99']
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%',
         startangle=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
df9=df8[df8['Updated On']=='06/07/2021']
```

```python
df9=df9[df9['State']!='India']
```

## State/Union Territories wise analysis of vaccination

```python
fig = plt.figure()
width = 0.4
ax1 = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax1.twinx() # Create another axes that shares the same x-axis as ax1.
ax = df9.plot(x="State", y=['Male(Individuals Vaccinated)','Female(Individuals Vaccinated)'],
              kind='bar',ax=ax1,position=1,width=width)
ax.get_yaxis().get_major_formatter().set_scientific(False)
df9.plot(x="State", y=['Transgender(Individuals Vaccinated)'],
              kind='bar',ax=ax2,position=0,width=0.2,color='green')
ax1.set_ylabel('Indivisual Vaccinate')
plt.rcParams['figure.figsize'] = [20, 15]
plt.show()
```

##### From graph it is clear that Uttar pradesh has highest male & transgender indivisuals who got vaccinated where as Madhya Pradesh has highest female indivisuals who have been vaccinated till 6th july 2021 

```python
# Total deaths till 7 july 2021 state wise
sns.set_theme(style="whitegrid")
ax = sns.barplot(x="State", y="Total Individuals Vaccinated", data=df9)
plt.xticks(rotation=90)
plt.show()
```

#### From above plot we can conclude that Uttar Pradesh has highest Indivisuals vaccinated follwed by Maharashatra and Rajasthan
